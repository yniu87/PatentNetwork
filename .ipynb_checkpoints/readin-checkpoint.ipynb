### merge with patents and keep only utility patents

```python
citation = citation.merge(patent, left_on = "patent_id", right_on = "id", how = 'left')
```

```python
citation.groupby('type').count()
```

# 89% of citations do not have a type..... impossible 

```python
citation[citation['type'].isnull()].shape[0] / citation.shape[0]
```

```python
citation = citation[~citation['type'].isin(['design','plant','reissue','statutory invention registration'])]
```

### add categories from NBER classification

```python
citation = citation.merge(nber[['patent_id','subcategory_id']], on = "patent_id", how = 'left')\
        .rename(columns={"subcategory_id": "patent_subcategory_id"})
```

```python
# 32% of patents don't have a subcategory_id 
citation[citation['patent_subcategory_id'].isnull()].shape[0] / citation.shape[0]
```

```python
citation = citation.merge(nber[['patent_id','subcategory_id']], left_on = "citation_id", right_on = "patent_id", \
                          how = 'left').rename(columns={"subcategory_id": "citation_subcategory_id"})
```

```python
# 14% of patents don't have a subcategory_id 
citation[citation['citation_subcategory_id'].isnull()].shape[0] / citation.shape[0]
```

### calculate directional citation flows

```python
list_subcat = list(citation['patent_subcategory_id'].dropna().unique())
```

```python
list_subcat.sort()
```

```python
df_citation = pd.DataFrame()

for patent_subcategory_id in list_subcat:
    temp = citation[citation['patent_subcategory_id'] == patent_subcategory_id].groupby("citation_subcategory_id").count()\
        [['patent_subcategory_id']].rename(columns={"patent_subcategory_id": patent_subcategory_id})
    df_citation = df_citation.merge(temp, left_index=True, right_index=True, how = "outer")
```
